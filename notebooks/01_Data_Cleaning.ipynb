```python
import pandas as pd
import numpy as np

print("Libraries Loaded Successfully!")
```

```python
orders = pd.read_csv("olist_orders_dataset.csv")
customers = pd.read_csv("olist_customers_dataset.csv")
order_items = pd.read_csv("olist_order_items_dataset.csv")
products = pd.read_csv("olist_products_dataset.csv")
payments = pd.read_csv("olist_order_payments_dataset.csv")
reviews = pd.read_csv("olist_order_reviews_dataset.csv")
```

```python
datasets = {
    "Orders": orders,
    "Customers": customers,
    "Order Items": order_items,
    "Products": products,
    "Payments": payments,
    "Reviews": reviews
}

for name, df in datasets.items():
    print("=" * 60)
    print(name)
    print("=" * 60)
    print("Shape:", df.shape)
    print("\nColumns:")
    print(df.columns.tolist())
    print("\nMissing Values:")
    print(df.isnull().sum())
    print("\n")
```

```python
for name, df in datasets.items():
    print("=" * 60)
    print(f"{name}")
    print("=" * 60)

    print("First 5 Rows:")
    display(df.head())

    print("\nShape:", df.shape)

    print("\nData Types:")
    print(df.dtypes)

    print("\nMissing Values:")
    print(df.isnull().sum())

    print("\nDuplicate Rows:", df.duplicated().sum())

    print("\n")
```

```python
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    orders[col] = pd.to_datetime(orders[col])

orders.info()
```

```python
orders["delivery_days"] = (
    orders["order_delivered_customer_date"]
    - orders["order_purchase_timestamp"]
).dt.days
```

```python
orders["late_delivery"] = (
    orders["order_delivered_customer_date"]
    > orders["order_estimated_delivery_date"]
)
```

```python
orders.head()
```

```python
How many orders were delivered late?
```

```python
late_orders = orders["late_delivery"].sum()

print(f"Late Orders: {late_orders}")
print(f"Percentage: {late_orders / len(orders) * 100:.2f}%")
```

Average Delivery Time

```python
print("Average Delivery Time:", orders["delivery_days"].mean())
```

Visualize Order Status

```python
import matplotlib.pyplot as plt

orders["order_status"].value_counts().plot(kind="bar")

plt.title("Order Status Distribution")
plt.xlabel("Order Status")
plt.ylabel("Count")

plt.show()
```

Merge Orders + Customers

```python
orders_customers = pd.merge(
    orders,
    customers,
    on="customer_id",
    how="left"
)

print(orders_customers.shape)
orders_customers.head()
```

```python
on="customer_id"
```

```python
how="left"
```

```python
orders_customers.info()
orders_customers.isnull().sum()
```

 Add Order Items

```python
master_df = pd.merge(
    orders_customers,
    order_items,
    on="order_id",
    how="left"
)

print(master_df.shape)
master_df.head()
```

Add Product Information

```python
master_df = pd.merge(
    master_df,
    products,
    on="product_id",
    how="left"
)

print(master_df.shape)
```

Step 5: Add Payment Data

```python
master_df = pd.merge(
    master_df,
    payments,
    on="order_id",
    how="left"
)
```

Step 6: Add Reviews

```python
master_df = pd.merge(
    master_df,
    reviews,
    on="order_id",
    how="left"
)

master_df.head()
```

```python
print(master_df.shape)

master_df.info()

master_df.isnull().sum()
```

Total Orders

```python
total_orders = master_df["order_id"].nunique()

print("Total Orders:", total_orders)
```

Total Revenue

```python
total_revenue = master_df["price"].sum()

print(f"Total Revenue: ₹{total_revenue:,.2f}")
```

Total Customers

```python
total_customers = master_df["customer_unique_id"].nunique()

print("Total Customers:", total_customers)
```

Average Order Value (AOV)

```python
average_order_value = total_revenue / total_orders

print(f"Average Order Value: ₹{average_order_value:.2f}")
```

Top 10 Product Categories

```python
master_df.groupby("product_category_name")["price"] \
         .sum() \
         .sort_values(ascending=False) \
         .head(10)
```

6️⃣ Top 10 States by Revenue

```python
master_df.groupby("customer_state")["price"] \
         .sum() \
         .sort_values(ascending=False)
```

# Exploratory Data Analysis (EDA)

Monthly Revenue Trend

```python
master_df["month"] = master_df["order_purchase_timestamp"].dt.to_period("M")

monthly_sales = (
    master_df.groupby("month")["price"]
    .sum()
)

monthly_sales
```

Visualize it

```python
import matplotlib.pyplot as plt

monthly_sales.plot(figsize=(12,5))

plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")

plt.grid(True)

plt.show()
```

Analysis 2

Top 10 States

```python
top_states = (
    master_df.groupby("customer_state")["price"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_states.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 States by Revenue")
plt.ylabel("Revenue")

plt.show()
```

Order Status Distribution

```python
master_df["order_status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(8,8)
)

plt.title("Order Status Distribution")

plt.show()
```

Delivery Days Distribution

```python
master_df["delivery_days"].hist(
    bins=30,
    figsize=(10,5)
)

plt.title("Delivery Days Distribution")

plt.xlabel("Days")

plt.ylabel("Orders")

plt.show()
```

Top Product Categories

```python
top_categories = (
    master_df.groupby("product_category_name")["price"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_categories.plot(kind="barh", figsize=(10,6))

plt.title("Top Product Categories")

plt.show()
```
